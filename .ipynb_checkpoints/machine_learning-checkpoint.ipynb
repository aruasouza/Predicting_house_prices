```python
#Função que retorna a chave do menor valor em um dicionário
#Foi utilizada na função que transforma os dados categóricos em números
def menor(dicionario):
    mi = 10**10
    chave = ''
    for key in dicionario:
        if dicionario[key] < mi:
            mi = dicionario[key]
            chave = key
    return chave
```

```python
#Função que recebe um dataframe e transforma os dados categóricos em números
def numerizar(df):
    for column in df:
        if type(df[column][0]) != np.int64:
            mapa = dict()
            for row in range(len(df)):
                alvo = df[column][row]
                if alvo in mapa:
                    mapa[alvo].append(df['Sale_Price'][row])
                else:
                    mapa[alvo] = [df['Sale_Price'][row]]
            for elemento in mapa:
                mapa[elemento] = sum(mapa[elemento]) / len(mapa[elemento])
            ordem = dict()
            i = 0
            while len(mapa) > 0:
                ordem[menor(mapa)] = i
                mapa.pop(menor(mapa))
                i += 1
            for i in range(len(df)):
                df[column][i] = ordem[df[column][i]]

#Função que multiplica os dados por um fator k para equilibrá-los
#Não altera a correlação que os dados têm com o preço              
def equalizar(df):
    for coluna in df:
        if coluna == 'Sale_Price':
            pass
        elif df[coluna][0] < 100:
            for i in range(len(df[coluna])):
                df[coluna][i] = df[coluna][i] * 100
        elif df[coluna][0] < 500:
            for i in range(len(df[coluna])):
                df[coluna][i] = df[coluna][i] * 10
        elif df[coluna][0] < 1900:
            for i in range(len(df[coluna])):
                df[coluna][i] = df[coluna][i] * 1
        else:
            for i in range(len(df[coluna])):
                df[coluna][i] = (df[coluna][i] - 1900) * 10

#Função que cria uma população inicial de indivíduos aleatórios
#Recebe como entrada o número de indivíduos e o número de genes
def criar_populacao(genes,individuos):
    populacao = []
    for _ in range(individuos):
        individuo = []
        for _ in range(genes):
            gene = random.randrange(0,10)
            individuo.append(gene)
        populacao.append(individuo)
    return populacao

#Função que calcula a aptidão de um dado indivíduo
def testar_aptidao(ind,df):
    global previsoes
    estimativas = []
    erros = []
    for i in range(len(df)):
        #print('imovel: ',i)
        soma = 0
        for j in range(len(ind)):
            soma += ind[j] * df.iloc[i][j]
        estimativas.append(soma)
        preco = df['Sale_Price'][i]
        erro = (preco - soma) / preco
        modulo_erro = (erro ** 2) ** 0.5
        #print('preço: ',preco)
        #print('chute: ',soma)
        #print('erro: ',modulo_erro)
        erros.append(modulo_erro)
    erro_medio = sum(erros) / len(erros)
    apt = 1 / erro_medio
    apt = (apt - 1) * 10
    #print('erro médio: ',erro_medio)
    #print('aptidão: ',apt)
    previsoes = estimativas
    return apt

#Função que calcula a correlação que as previsões de dado indivíduo têm com o target
def testar_aptidao2(ind,df):
    global previsoes
    alvos = list(df['Sale_Price'])
    estimativas = []
    for i in range(len(df)):
        soma = 0
        for j in range(len(ind)):
            soma += ind[j] * df.iloc[i][j]
        estimativas.append(soma)
    data = {'alvos':alvos,'estimativas':estimativas}
    result = pd.DataFrame(data)
    matriz_corr = result.corr()
    #print(matriz_corr)
    corr = matriz_corr['alvos'][1]
    apt = corr * 100 - 80
    #print(estimativas)
    #print(corr)
    previsoes = estimativas
    #return apt
    return corr

#Função que testa a aptidão de toda uma população
def testar_pop(pop,df):
    aptidoes = []
    for ind in pop:
        aptidoes.append(testar_aptidao(ind,df))
    return aptidoes

#Função que cria uma nova geração de indivíduos com base nos resultados da geração anterior
def nova_geracao(geracao,aptidoes):
    apt_media = sum(aptidoes) / len(aptidoes)
    #print('aptidao media: ',apt_media)
    #print('maior aptidao: ',max(aptidoes))
    #print('aptidões: ',aptidoes)
    pesos = []
    for valor in aptidoes:
        if valor > 0:
            pesos.append(valor)
        else:
            pesos.append(0)
    #print('pesos: ',pesos)
    populacao = []
    for _ in range(len(geracao)):
        ind = []
        pai = random.choices(geracao,pesos)[0]
        #print('pai: ',pai)
        for gene in pai:
            #ind.append(int(random.triangular(-10,11,gene)))
            novo_gene = round(np.random.normal(gene,1,1)[0],0)
            if novo_gene < 0:
                novo_gene = 0
            ind.append(novo_gene)
        #print('ind',ind)
        populacao.append(ind)
    #print('população: ',populacao)
    return populacao

#Função que testa a eficiência de um dado indivíduo para uma dada margem de erro
def teste_eficiencia(ind,df,margem):
    acertos = 0
    tentativas = 0
    for i in range(len(df)):
        #print('imovel: ',i)
        soma = 0
        for j in range(len(ind)):
            soma += ind[j] * df.iloc[i][j]
        estimativas.append(soma)
        preco = df['Sale_Price'][i]
        erro = (preco - soma) / preco
        modulo_erro = (erro ** 2) ** 0.5
        if modulo_erro <= margem:
            acertos += 1
        tentativas += 1
    return acertos / tentativas
```

```python
#É definida a classe Tabela, utilizada para inserir o dataframe em uma tabela de uma database MySQL
#Ao mesmo tempo é criado um objeto que permite que esses dados sejam facilmente extraídos
class Tabela:
    #Função construtora do objeto. Caso a tabela já exista na database os dados são selecionados
    #e armazenados em self.data se a tabela estiver vazia os dados são inseridos
    #Se a tabela não existir na database ela é criada e os dados são inseridos
    def __init__(self,cursor,nome,df):
        self.atributos = df.columns.values.tolist()
        self.nome = nome
        self.cursor = cursor
        try:
            comando = 'SELECT * FROM ' + self.nome + ';'
            self.cursor.execute(comando)
            self.data = self.cursor.fetchall()
            if self.data == []:
                empty = True
            else:
                empty = False
        except mysql.ProgrammingError:
            empty = True
            list_tipos = []
            for item in self.atributos:
                if type(df[item][0]) == np.int64:
                    list_tipos.append('INT')
                elif type(df[item][0]) == np.float64:
                    list_tipos.append('DOUBLE')
                elif type(df[item][0]) == str:
                    list_tipos.append('VARCHAR(50)')
            atributos_str = '('
            for i in range(len(self.atributos)):
                atributos_str += self.atributos[i] + ' ' + list_tipos[i] + ','
            atributos_str = atributos_str[:-1] + ')'
            comando = 'CREATE TABLE ' + nome + ' ' + atributos_str + ';'
            #print(comando)
            self.cursor.execute(comando)
        finally:
            if empty:
                for i in range(len(df)):
                    lista_dados = list(df.iloc[i])
                    nomes = '('
                    for nome in self.atributos:
                        nomes += nome + ','
                    nomes = nomes[:-1] + ')'
                    lista = '('
                    for dado in lista_dados:
                        if type(dado) != str:
                            lista += str(dado) + ','
                        else:
                            dado = '"' + dado + '"'
                            lista += dado + ','
                    lista = lista[:-1] + ')'
                    comando = 'INSERT INTO ' + self.nome + ' ' + nomes + ' VALUES ' + lista + ';'
                    #print(comando)
                    self.cursor.execute(comando)
                self.select()
    #Função que apaga os dados da tabela    
    def clear(self):
        comando = 'DELETE FROM ' + self.nome + ';'
        self.cursor.execute(comando)
        self.select()
    #Função que retorna os dados da tabela como uma lista de tuplas    
    def select(self):
        comando = 'SELECT * FROM ' + self.nome + ';'
        self.cursor.execute(comando)
        self.data = self.cursor.fetchall()
        return self.data
    #Função que retorna uma coluna de uma tabela como uma lista    
    def coluna(self,nome):
        comando = 'SELECT ' + nome + ' FROM ' + self.nome + ';'
        self.cursor.execute(comando)
        col = self.cursor.fetchall()
        lista = list()
        for i in range(len(col)):
            lista.append(col[i][0])
        return lista
    #Função que retorna várias colunas como um dicionário de listas
    def get_colunas(self,lista):
        mapa = dict()
        for item in lista:
            mapa[item] = self.coluna(item)
        return mapa
    #Função que retorna todos od dados como um dataframe
    def get_dataframe(self):
        mapa = dict()
        for item in self.atributos:
            mapa[item] = self.coluna(item)
        df = pd.DataFrame(mapa)
        return df
```

```python
#A tabela é criada na database e os dados são inseridos
my_sql_table = Tabela(cursor,'imoveis_table',df)
```
